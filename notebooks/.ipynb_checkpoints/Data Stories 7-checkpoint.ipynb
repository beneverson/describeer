```python
beer_df = pd.DataFrame.from_csv(allbeers_csv_path) # load all the beer reviews
```

```python
beer_df.columns
```

```python
styles_grouped = beer_df.groupby('beer/style')
```

```python
styles = []
number_of_reviews = []
style_and_review = []
for name, group in styles_grouped:
    styles.append(name)
    number_of_reviews.append(group.shape[0])
    style_and_review.append((name, group.shape[0]))
```

```python
beer_df['review/profileName'].nunique()
```

```python
style_and_review.sort(key=lambda tup: tup[1], reverse=True)
```

```python
style_and_review[:10]
```

```python
style_numreviews_df = pd.DataFrame()
style_numreviews_df['style'] = styles
style_numreviews_df['num_reviews'] = number_of_reviews
style_numreviews_df.sort(columns='num_reviews', inplace=True, ascending=False)
style_numreviews_df.index = range(1,len(style_numreviews_df) + 1)
```

```python
#style_numreviews_df.set_index('style', inplace=True)
```

```python
%matplotlib inline
ax = style_numreviews_df.loc[:7].plot(x='style', y='num_reviews', kind='bar', legend=None)
ax.set_ylabel('Number of Reviews', fontsize=20)
ax.set_xlabel('Beer Style', fontsize=20)
fig = ax.get_figure()
fig.savefig("ReviewsByStyleTopTenVert.png", bbox_inches="tight")
```

```python
# overall score graph
ax = beer_df['review/overall'].hist(bins=8)
ax.set_xlabel('Overall Rating', fontsize=20)
ax.set_ylabel('Number of Reviews', fontsize=20)
ax.set_ylim([0, 350000])
fig = ax.get_figure()
fig.savefig("OverallScoreDist.png", bbox_inches="tight")
```

```python
beers_grouped = beer_df.groupby('beer/name')
beer_name = []
number_of_reviews_by_beer = []
for name, group in beers_grouped:
    beer_name.append(name)
    number_of_reviews_by_beer.append(group.shape[0])
```

```python
beer_numreviews_df = pd.DataFrame()
beer_numreviews_df['beer_name'] = beer_name
beer_numreviews_df['num_reviews'] = number_of_reviews_by_beer
beer_numreviews_df.sort(columns='num_reviews', inplace=True, ascending=False)
beer_numreviews_df.index = range(1,len(beer_numreviews_df) + 1)
```

```python
ax = beer_numreviews_df.loc[:7].plot(x='beer_name', y='num_reviews', kind='bar', legend=None)
ax.set_xlabel('Beer Name', fontsize=20)
ax.set_ylabel('Number of Reviews', fontsize=20)
fig = ax.get_figure()
fig.savefig("ReviewsByBeerTopTenVert.png", bbox_inches="tight")
```

```python
beers_by_number_of_reviews = beer_numreviews_df.groupby('num_reviews')
number_of_beers = []
number_of_reviews = []
for name, group in beers_by_number_of_reviews:
    number_of_reviews.append(int(name))
    number_of_beers.append(group.shape[0])
```

```python
num_beers_num_reviews_df = pd.DataFrame()
num_beers_num_reviews_df['num_beers'] = number_of_beers
num_beers_num_reviews_df['num_reviews'] = number_of_reviews
```

```python
ax = beer_numreviews_df['num_reviews'].hist(bins=15)
ax.set_xlabel('Number of Reviews', fontsize=20)
ax.set_ylabel('Number of Beers', fontsize=20)
ax.set_xlim([225, 3500])
fig = ax.get_figure()
fig.savefig("BeersByReviewDist.png", bbox_inches="tight")
```

```python
beer_df = beer_df[pd.notnull(beer_df['review/text'])]
```

```python
beer_df['review/length'] = beer_df['review/text'].apply(lambda x: len(x.split()))
```

```python
ax = beer_df['review/length'].hist(bins=50)
ax.set_xlabel('Length of Review (Words)', fontsize=20)
ax.set_ylabel('Number of Beers', fontsize=20)
ax.set_xlim([0, 500])
fig = ax.get_figure()
fig.savefig("BeersByReviewLengthDist.png", bbox_inches="tight")
```

```python
from gensim.models import Doc2Vec
import cPickle as pickle
model_path = "../app/describeer/ReviewIndividualSentencesMultiLabelFixedAlphaTrained1x.doc2vec"
namelist_pickle_path = "../app/describeer/pickles/namelist.pickle"
stylelist_pickle_path = "../app/describeer/pickles/stylelist.pickle"
nametostyle_pickle_path = "../app/describeer/pickles/nametostyle.pickle"
```

```python
model = Doc2Vec.load(model_path) # load the model
all_beernames = pickle.load(open(namelist_pickle_path, 'rb'))
all_beerstyles = pickle.load(open(stylelist_pickle_path, 'rb'))
name_to_style = pickle.load(open(nametostyle_pickle_path, 'rb'))
```

```python
# collect the vectors for all names
beername_vectors = []
for beername in all_beernames:
    beername_vectors.append(model[beername])
# and all styles   
beerstyle_vectors = []
for beerstyle in all_beerstyles:
    beerstyle_vectors.append(model[beerstyle])
```

```python
# run pca/lda on the vectors
import numpy as np
from sklearn.decomposition import PCA
from sklearn.lda import LDA
pca = PCA(n_components=3)
lda = LDA(n_components=3)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)

# beer style vectors
styles_pca = pca.fit_transform(beerstyle_vectors)
plt.figure()
ax.scatter(styles_pca[:,0], styles_pca[:,1], styles_pca[:,2])
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)

# beer name vectors
names_pca = pca.fit_transform(beername_vectors)
ax.scatter(names_pca[:,0], names_pca[:,1], names_pca[:,2])
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(all_beerstyles)

#fig = plt.figure(1, figsize=(8, 6))
#ax = Axes3D(fig, elev=-150, azim=110)

stylelabels = []
for beername in all_beernames:
    stylelabels.append(le.transform(name_to_style[beername]))

# beer name vectors
names_pca = pca.fit_transform(beername_vectors)
plt.figure()
plt.scatter(names_pca[:,0], names_pca[:,1], c=stylelabels)
```

```python
names_lda = lda.fit_transform(beername_vectors, y=stylelabels)
plt.figure()
plt.scatter(names_lda[:,0], names_lda[:,1], c=stylelabels)
```

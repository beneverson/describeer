```python
styles_grouped = beer_df.groupby('beer/style')
```

```python
unique_beers_per_style = pd.DataFrame(styles_grouped['beer/name'].nunique()).sort(columns='beer/name')
```

```python
numreviews = []
stylename = []
for name, group in styles_grouped:
    numreviews.append(group.shape[0])
    stylename.append(name)
reviews_per_style = pd.DataFrame()
reviews_per_style['style'] = stylename
reviews_per_style['count'] = numreviews
reviews_per_style.set_index('style', inplace=True)
reviews_per_style.sort(inplace=True, columns='count')
```

```python
grouped_by_username = beer_df.groupby('review/profileName')
```

```python
username = []
count = []
for name, group in grouped_by_username:
    username.append(name)
    count.append(group.shape[0])
username_and_reviewcount_df = pd.DataFrame()
username_and_reviewcount_df['username'] = username
username_and_reviewcount_df['count'] = count
```

```python
grouped_by_reviewcount = username_and_reviewcount_df.groupby('count')
```

```python
unique_reviewers_by_count = pd.DataFrame(grouped_by_reviewcount['username'].nunique())
```

```python
ax = unique_reviewers_by_count.plot(legend=False, title="Reviews Per User")
ax.set_ylim([0, 750])
ax.set_xlim([0, 500])
ax.set_xlabel("Number of Reviews")
ax.set_ylabel("Number of Users")
fig = ax.get_figure()
fig.set_size_inches(10, 10)
fig.savefig("NumReviewsDistribution.png")
```
